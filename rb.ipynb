````python
def generate_blog_v2(title,next_title):
  # title = "Python - Everything starts from here. Hello World in Python"
  prompt = f"""Write a blog about {title} with the following guidelines:

IMPORTANT:-
1. Use a conversational, human-like tone.
2. welcome readers before starting the content in short way possible.
3. create content close to title. dont create additional content.
4. Enclose the blog title in <title> tags.
5. you should Use <h> tags for all headings (not <h1>, <h2>, etc.).
6. Write paragraphs enclosed in  <p> tags.
7. Include example code if needed, and enclose it in <code> tags.
8. End with an engaging message encouraging readers to return.
9. Tease the next blog title: {next_title}.
10. Avoid repeating same examples across blogs.
11. No nested or additional HTML tags should be used.
12. ensure all contents are enclosed in tags mentioned above. No text should be left without tags.

make sure above guidlines are folllowed strictly
              """
  messages = [
    {"role": "system", "content": """You are a helpful assistant to write tech blogs.
                                      there should be no sensitive content or no harmful words.
                                      """},
    {"role": "user", "content": prompt}
  ]
  chat_template_prompt = tokenizer.apply_chat_template(
      messages,
      return_tensors="pt"
  )
  model_inputs = chat_template_prompt.to("cuda")
  generated_ids = model.generate(model_inputs, max_new_tokens=1500, do_sample=True)
  decoded = tokenizer.batch_decode(generated_ids)
  output = decoded[0].split("[/INST]")[1].replace("</s>","")
  # print(decoded[0])
  return output

````
